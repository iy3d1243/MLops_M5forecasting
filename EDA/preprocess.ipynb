```python
#downcast for less memory usage 
cols_int16={
    'id': 'int16',
    'item_id': 'int16',
    'dept_id': 'int16',
    'cat_id': 'int16',
    'store_id': 'int16',
    'state_id': 'int16',
    'sales': 'int16',
    'wm_yr_wk': 'int16',
    'month': 'int16',
    'year': 'int16',
    'd': 'int16',
    'event_name_1': 'int16',
    'event_type_1': 'int16',
    'event_name_2': 'int16',
    'event_type_2': 'int16',
    'snap_CA': 'int16',
    'snap_TX': 'int16',
    'snap_WI': 'int16',
    'day': 'int16'
    }
```

```python
df=pd.read_csv('final.csv',dtype=cols_int16)
```

```python
col_ord=['id','item_id','dept_id','cat_id','store_id','state_id','wm_yr_wk','day','month','year','d','event_name_1','event_type_1','event_name_2','event_type_2','snap_CA','snap_TX','snap_WI','weekday_cos','weekday_sin','sales','sell_price']
```

```python
df=df[col_ord]
```

```python
df.rename(columns={'day_y':'day'},inplace=True)
```
